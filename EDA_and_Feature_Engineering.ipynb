# Differentials
We have A_Wins and H_Wins columns, but that number will be a whole lot different in September than it is in May. For example, 50 wins in June is really good, but 50 wins in September is really bad. So we'll derive win differential variables.<br>

We also should do the same with Pythagorean Wins.

```python
main_df['A_Team_Wins'].max(), main_df['A_Team_Wins'].min(), main_df['H_Team_Wins'].max(), main_df['H_Team_Wins'].min()
```

```python
main_df['A_Win_Diff'] = main_df['A_Team_Wins'] - main_df['H_Team_Wins']
```

```python
main_df['H_Win_Diff'] = main_df['H_Team_Wins'] - main_df['A_Team_Wins']
```

```python
main_df['A_Py_Diff'] = main_df['A_Py_Wins'] - main_df['H_Py_Wins']
```

```python
main_df['H_Py_Diff'] = main_df['H_Py_Wins'] - main_df['A_Py_Wins']
```

```python
main_df = main_df.drop(columns = ['A_Team_Wins', 'H_Team_Wins', 'A_Py_Wins', 'H_Py_Wins'])
```

So that the target variable can have a binary value of 1 or 0, we'll base it on whether the home team wins.

```python
main_df['Home_Win'] = np.where(main_df['Winner'] == 'H', 1, 0)
```

# Drop redundant columns
There are a few columns that are now redundant. For example, the win differential and the Pythagorean Win Differentials will be opposite numbers adding up to zero. So we can drop the A_Win_Diff, for example, because we know if H_Win_Diff is 9, that means the away team has nine less wins than the home team and A_Win_Diff would be -9. We'll run a couple of lines of code that demonstrate this.

```python
main_df['A_Win_Diff'].sum(), main_df['H_Win_Diff'].sum()
```

```python
main_df['A_Py_Diff'].sum(), main_df['H_Py_Diff'].sum()
```

```python
main_df = main_df.drop(columns = ['A_Win_Diff', 'A_Py_Diff'])
```

```python
main_df = main_df.rename({'H_Win_Diff': 'Win_Diff', 'H_Py_Diff': 'Py_Diff'})
```

# A differential of a differential
The run differentials for the home team and away team aren't symmetrical in a single game. A home team's run differential of plus-25 might show superiority in one game if the away team's run differential is minus-3, but on another day it could mean an  even matchup if the away team's run differential is also plus-25. So we'll subtract the visiting team's run differential from the home team's run differential and create a single 'Run_Diff' variable.<br>

We'll also do the same with wins in last 10 and 30 games.

```python
main_df['Run_Diff'] = main_df['H_Run_Diff'] - main_df['A_Run_Diff'] 
```

```python
main_df = main_df.drop(columns = ['H_Run_Diff', 'A_Run_Diff'])
```

```python
main_df['W_L10_Diff'] = main_df['H_Wins_L10'] - main_df['A_Wins_L10']
main_df['W_L30_Diff'] = main_df['H_Wins_L30'] - main_df['A_Wins_L30']
```

```python
main_df = main_df.drop(columns = ['H_Wins_L10', 'A_Wins_L10', 'H_Wins_L30', 'A_Wins_L30'])
```

# BABIP
Now that we have single variables that indicate the difference in wins, Pythagorean Wins, wins in last 10 and wins in last 30 between the home and away teams, we should do that for the rest of the variables.<br>

For example, we can create a variable that indicates the home team's OPS minus the visiting team's OPS. After all, the point of this model is to compare one team against another on a particular day. In most cases, we'll need to just compare hitting data to hitting data and pitching data to pitching data. But there is one variable that allows us to compare pitching to hitting.<br>

We have BABIP, batting average on balls in play, for each offense as well as the starting pitcher and bullpen. So we'll derive the home team's BABIP minus the away team's SP and bullpen BABIPs and the away team's BABIP minus the home team's SP and bullpen BABIPs.

```python
main_df['BABIP_vs_SP'] = main_df['H_BABIP'] - main_df['A_BABIP_SP']
main_df['BABIP_vs_BP'] = main_df['H_BABIP'] - main_df['A_BABIP_BP']
main_df['BABIP_SP_vs_Hit'] = main_df['H_BABIP_SP'] - main_df['A_BABIP']
main_df['BABIP_BP_vs_Hit'] = main_df['H_BABIP_BP'] - main_df['A_BABIP']
```

```python
main_df = main_df.drop(columns = ['A_BABIP', 'A_BABIP_SP', 'A_BABIP_BP', 'H_BABIP', 'H_BABIP_SP', 'H_BABIP_BP'])
```

```python
main_df['Avg_Outs_Diff'] = main_df['H_Avg_Outs'] - main_df['A_Avg_Outs']
main_df['FIP_Diff'] = main_df['H_FIP'] - main_df['A_FIP']
main_df['FIP_BP_Diff'] = main_df['H_FIP_BP'] - main_df['A_FIP_BP']
main_df['ISO_Diff'] = main_df['H_ISO'] - main_df['A_ISO']
main_df['OPS_Diff'] = main_df['H_OPS'] - main_df['A_OPS']
main_df['RS/9_Diff'] = main_df['H_RS/9'] - main_df['A_RS/9']
main_df['SIERA_Diff'] = main_df['H_SIERA'] - main_df['A_SIERA']
main_df['SIERA_BP_Diff'] = main_df['H_SIERA_BP'] - main_df['A_SIERA_BP']
main_df['WAR_Diff'] = main_df['H_WAR'] - main_df['A_WAR']
main_df['WHIP_Diff'] = main_df['H_WHIP'] - main_df['A_WHIP']
main_df['WHIP_BP_Diff'] = main_df['H_WHIP_BP'] - main_df['A_WHIP_BP']
main_df['wOBA_Diff'] = main_df['H_wOBA'] - main_df['A_wOBA']
main_df['xFIP_Diff'] = main_df['H_xFIP'] - main_df['A_xFIP']
main_df['xFIP_BP_Diff'] = main_df['H_xFIP_BP'] - main_df['A_xFIP_BP']
```

```python
main_df = main_df.drop(columns = ['A_Avg_Outs', 'H_Avg_Outs', 'A_FIP', 'H_FIP', 'A_FIP_BP', 'H_FIP_BP', 'A_ISO', 'H_ISO',\
                                 'A_OPS', 'H_OPS', 'A_RS/9', 'H_RS/9', 'A_SIERA', 'H_SIERA', 'A_SIERA_BP', 'H_SIERA_BP',\
                                 'A_WAR', 'H_WAR', 'A_WHIP', 'H_WHIP', 'A_WHIP_BP', 'H_WHIP_BP', 'A_wOBA', 'H_wOBA',\
                                 'A_xFIP', 'H_xFIP', 'A_xFIP_BP', 'H_xFIP_BP'])
```
