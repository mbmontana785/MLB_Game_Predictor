# Last 10
Now we need a wins-in-last-10-games variable. This for loop creates a temporary dataframe for each team and subtracts Team_Wins from 10 rows before from Team_Wins in the current row of the loop. We need to reset the index of each temp_df so we can use the index when we use .loc.<br>

Then we concat the list of temp_dfs and sort by Date and game_id.

```python
list_of_dfs = []
for team in wins_21['Team'].unique():
    temp_df = wins_21[wins_21['Team'] == team]
    temp_df = temp_df.reset_index(drop=True)
    temp_df['Wins_L10'] = 0
    for i in range(10):
        temp_df.loc[i, 'Wins_L10'] = temp_df.loc[i, 'Team_Wins']
    for j in range(10, (len(temp_df))):
        temp_df.loc[j, 'Wins_L10'] = temp_df.loc[j, 'Team_Wins'] - temp_df.loc[j-10, 'Team_Wins']
    list_of_dfs.append(temp_df)
        
```

```python
wins_21 = pd.concat(list_of_dfs).sort_values(['Date', 'game_id', 'Host'])
```

# Last 30

# Pythagorean wins
Now let's add a Pythagorean Wins variable. According to <a href='https://www.mlb.com/glossary/advanced-stats/pythagorean-winning-percentage'>MLB.com</a>, this is a formula created by noted baseball statistician Bill James to measure how many games a team "should have" won, based on runs scored and runs allowed.<br>

**Formula:** Runs Scored to the 1.83 power/((Runs Scored to the 1.83 power) + (Runs Allowed to the 1.83 power))<br>

We'll start by creating a variable that counts each team's games played, then cumulative Runs_Scored and Runs_Allowed variables.<br>

```python
wins_21 = wins_21.reset_index(drop=True)
```

```python
wins_21['Games_Played'] = wins_21.groupby('Team').cumcount() + 1
```

```python
wins_21.tail(10)
```

```python
wins_21['Opp_Score'] = 0
```

```python
for i in range(len(wins_21)):
    if i % 2 == 0:
        wins_21.loc[i, 'Opp_Score'] = wins_21.loc[i+1, 'Score']
    else:
        wins_21.loc[i, 'Opp_Score'] = wins_21.loc[i-1, 'Score']
            
    
```

```python
wins_21.head(10)
```

```python
wins_21['Runs_Scored'] = wins_21.groupby('Team')['Score'].cumsum()
```

```python
wins_21.tail(10)
```

```python
wins_21['Runs_Allowed'] = wins_21.groupby('Team')['Opp_Score'].cumsum()
```

```python
wins_21.tail(10)
```

```python
wins_21['Py_Wins'] = np.round(wins_21['Runs_Scored']**1.83/((wins_21['Runs_Scored']**1.83)+(wins_21['Runs_Allowed']**1.83))\
*wins_21['Games_Played'], 0)
```

```python
wins_21.iloc[1000:1005, :]
```

Now our runs-differential feature

```python
wins_21['Run_Diff'] = wins_21['Runs_Scored'] - wins_21['Runs_Allowed']
```

# Columns to shift
We'll create a list of columns to shift. Then once again we'll loop through all 30 teams and shift these columns down one so that we have this data **going into** the next game so that it's predictive.

```python
cols_to_shift = ['Team_Wins', 'Wins_L10', 'Wins_L30', 'Run_Diff', 'Py_Wins']
```

```python
chatGPT("How do I shift the values in a column down one in pandas?")
```

```python
wins_21.info()
```

```python
list_of_dfs = []
for team in wins_21['Team'].unique():
    temp_df = wins_21[wins_21['Team'] == team]
    temp_df = temp_df.reset_index(drop=True)
    for col in cols_to_shift:
        temp_df[col] = temp_df[col].shift(periods = 1)
    list_of_dfs.append(temp_df)
```

```python
wins_21 = pd.concat(list_of_dfs).sort_values(by = ['Date', 'game_id', 'Host'])
```
